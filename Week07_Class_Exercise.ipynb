# Construct a function for extract individual pages

```
def extract_info(url):
  print(url)
  html = urlopen(url)
  bs = BeautifulSoup(html, "html.parser")
  # extract information
  name_tags = bs.select('.primary-info')
  mp_names = [item.get_text().strip() for item in name_tags]
  party_names = [item.get_text().strip() for item in bs.select('.secondary-info')]
  const_names = [item.get_text().strip() for item in bs.select('.indicator-label')]
  full_url = ["https://members.parliament.uk" + item['href'] for item in bs.select('.card-member')]
  # creating dataframe
  df_mps = pd.DataFrame(mp_names, columns = ['Name'])
  df_mps['Const'] = const_names
  df_mps['Party'] = party_names
  df_mps['Ind_page'] = full_url
  sleep(1)
  
  return(df_mps)
```

```
extract_info("https://members.parliament.uk/members/commons")
```

# Generate a list of URLs

```
list_urls = ["https://members.parliament.uk/members/commons?page=" + str(i) for i in range(1, 34)]
```

```
extract_info(list_urls[2])
```

# Extract all pages

## Using for loop

```
df_all_mps = pd.DataFrame()

for url in list_urls:
  df_all_mps = df_all_mps.append(extract_info(url), ignore_index = True)
```

```
df_all_mps
```

## Using list comprehension

```
list_dfs = [extract_info(current_url) for current_url in list_urls]
list_dfs
```

```
df_all_mps_2 = pd.concat(list_dfs, ignore_index=True)
df_all_mps_2
```

## Using map()

```
list_dfs_2 = list(map(extract_info, list_urls))
```

```
df_all_mps_3 = pd.concat(list_dfs_2, ignore_index=True)
df_all_mps_3.head()
```

# Visit individual page and get some information

For example:

- https://members.parliament.uk/member/1586/contact

```
bs = BeautifulSoup(urlopen('https://members.parliament.uk/member/1586/contact'), 'html.parser')
```

```
def extract_more(url):
  print("fetching: " + url) # this line give you the idea of the progress
  bs = BeautifulSoup(urlopen(url), 'html.parser')
  tw_links = bs.select('a[href*="twitter"]') # find a-tags with a string "twitter" in href 
  tw_url = tw_links[0]['href'] if (len(tw_links) > 0) else None
  tel_links = bs.select('a[href*="tel:"]')
  tel_number = tel_links[0]['href'] if (len(tel_links) > 0) else None
  email_links = bs.select('a[href*="@"]')
  email_address = email_links[0]['href'] if (len(email_links) > 0) else None
  out = {"Ind_page": url, "twitter": tw_url, "tel": tel_number, "email": email_address} # this part I changes to keep "Ind_page" filed for mergoing with original dataset
  sleep(1) # with 1 sec break, the download works smoothly, but it will take more than 10 min to run the code below (650 secs )
  return(pd.DataFrame(out, index = [0]))
```

```
list_df_more = [extract_more(current_url) for current_url in df_all_mps_3['Ind_page']]
```

```
df_additional = pd.concat(list_df_more, ignore_index=True)
df_additional
```

Cleaning up the text 

```
df_additional['tel'] = df_additional['tel'].str.extract(r'([0-9 ]+)') # find sequence of numbers or spaces
df_additional['email'] = df_additional['email'].str.extract(r'mailto:(\S+)') # remove "mailto:"
```

```
df_final = df_all_mps_3.merge(df_additional, on = "Ind_page")
```
